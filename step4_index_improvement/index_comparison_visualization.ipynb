# Index Optimization Performance Comparison
This notebook visualizes the performance improvement across **No Index**, **Index V1**, and **Index V2**.
Data is based on the 100% dataset scale.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the provided table
queries = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7-a', 'Q7-b', 'Q8-opt']

no_index = [5.502, 35, 4.178, 49, 18, 15, 180, 14, 88]
index_v1 = [3.646, 30, 4.213, 46, 29, 9.967, 177, 10, 89]
index_v2 = [0.966, 31, 2.217, 32, 7.911, 3.871, 163, 0.031, 84]

v2_improvement = [82.44, 11.43, 46.94, 34.69, 56.05, 74.19, 9.44, 99.78, 4.55]

x = np.arange(len(queries))
width = 0.25

# Style
plt.rcParams.update({
    'font.family': 'sans-serif',
    'axes.spines.top': False,
    'axes.spines.right': False,
})

colors = {'No_Index': '#EF476F', 'Index_v1': '#FFD166', 'Index_v2': '#06D6A0', 'Improvement': '#118AB2'}
```

### 1. Grouped Bar Chart (Linear Scale)
Direct comparison of runtime across all queries.

```python
fig, ax = plt.subplots(figsize=(14, 6))

bars1 = ax.bar(x - width, no_index, width, label='No Index', color=colors['No_Index'], edgecolor='white')
bars2 = ax.bar(x, index_v1, width, label='Index v1', color=colors['Index_v1'], edgecolor='white')
bars3 = ax.bar(x + width, index_v2, width, label='Index v2', color=colors['Index_v2'], edgecolor='white')

ax.set_xticks(x)
ax.set_xticklabels(queries, fontsize=11)
ax.set_ylabel('Runtime (s)', fontsize=12)
ax.set_title('Query Runtime Comparison: No Index vs Index v1 vs Index v2', fontsize=16, pad=15, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(axis='y', linestyle='--', alpha=0.4)

# Add values on top of Index v2 bars for clarity
for bar in bars3:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, h + 2,
            f'{h:g}', ha='center', va='bottom', fontsize=9, color='#333', fontweight='bold')

plt.tight_layout()
plt.savefig('index_comparison_linear_bar.png', dpi=150)
plt.show()
```

### 2. V2 Improvement Percentage Chart
Highlights which queries benefited the most from the V2 index optimizations (Trigram, Composite, etc.).

```python
fig, ax = plt.subplots(figsize=(12, 5))

bars = ax.bar(x, v2_improvement, width=0.5, color=colors['Improvement'], alpha=0.9)

# Highlight high improvements (>50%)
for i, bar in enumerate(bars):
    if v2_improvement[i] > 50:
        bar.set_color('#06D6A0')

ax.set_xticks(x)
ax.set_xticklabels(queries, fontsize=11)
ax.set_ylabel('Improvement vs No Index (%)', fontsize=12)
ax.set_title('Index v2 Performance Improvement (%)', fontsize=16, pad=15, fontweight='bold')
ax.set_ylim(0, 110)
ax.grid(axis='y', linestyle='--', alpha=0.4)

for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, h + 2,
            f'{h}%', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('index_v2_improvement_bar.png', dpi=150)
plt.show()
```
